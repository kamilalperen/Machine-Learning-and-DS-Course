```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = np.linspace(0,10,11)
b = a**4
```

```python
x = np.arange(0,10)
y = 2*x
```

```python
fig, axes = plt.subplots()
```

```python
fig, axes = plt.subplots()
axes.plot(x,y)
```

```python
fig, axes = plt.subplots(nrows=2, ncols=1)
```

```python
fig, axes = plt.subplots(nrows=2, ncols=1)
axes[1].plot(x,y)
axes[0].plot(a,b)
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)
axes[1].plot(x,y)
axes[0].plot(a,b)
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2) # we have 2 dimensional graphs. 
axes[1].plot(x,y)
axes[0].plot(a,b)
```

```python
axes.shape
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2) 
axes[1][0].plot(x,y)
axes[0][1].plot(a,b)
```

```python
fig, axes = plt.subplots(nrows=3, ncols=1) # 1 dimensional. 

for ax in axes:
    ax.plot(x,y)
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2) 
axes[1][0].plot(x,y)
axes[0][1].plot(a,b)

plt.tight_layout()  # Adjusts the distance between graphics.
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2) 

axes[0][0].plot(x,y)

axes[0][1].plot(x,y)
axes[0][1].set_xlim(2,6)

axes[1][0].plot(x,y)
axes[1][0].set_ylabel('Y Label 1,0')

axes[1][1].plot(a,b)
axes[1][1].set_title('TITLE 1,1', fontsize=8)


fig.suptitle('Figure Level', fontsize=16)

fig.subplots_adjust(wspace=1, hspace=0.5)   # we adjust the distances between graphics.


fig.savefig('new_subplots.png', bbox_inches='tight')
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2, figsize = (10,5), dpi = 100) # The difference between figsize and dpi is important.

axes[0][0].plot(x,y)

axes[0][1].plot(x,y)
axes[0][1].set_xlim(2,6)

axes[1][0].plot(x,y)
axes[1][0].set_ylabel('Y Label 1,0')

axes[1][1].plot(a,b)
axes[1][1].set_title('TITLE 1,1', fontsize=8)


fig.suptitle('Figure Level', fontsize=16)

fig.subplots_adjust(wspace=1, hspace=0.5)   # we adjust the distances between graphics.


fig.savefig('new_subplots.png', bbox_inches='tight')
```

```python
x = np.linspace(0,10,11)
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,x)

ax.plot(x,x**2)
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,x, label='X vs X')

ax.plot(x,x**2, label= 'X vs X**2')

ax.legend(loc=4)      # loc: where locate the legend box. We can locate where we want (1.1, 1).
# The ax.legend() function adds a legend box that shows what data the lines on the chart represent, 
# ensuring that the labels specified in the label parameter appear on the chart.
```

```python
x =np.linspace(0,10,11)
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,x)
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,x,color='purple') # we can use also rgb HEX Code.
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,x, color='#c9520c')
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,x, color='#c9520c', label='X vs X')

ax.plot(x,x+1, color='green', label='X vs X+1')

ax.legend()
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,x, color='#c9520c', label='X vs X', lw=6) # line width (lw = linewidth).

ax.plot(x,x+1, color='green', label='X vs X+1')

ax.legend()
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,x, color='#c9520c', label='X vs X', lw=6,) # line width (lw = linewidth).

ax.plot(x,x+1, color='green', label='X vs X+1', ls='-.') # linestyle=ls

ax.legend()
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

lines = ax.plot(x,x, color='#c9520c', lw=6,) # line width (lw = linewidth).

#ax.plot(x,x+1, color='green', label='X vs X+1', ls='-.') # linestyle=ls
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,x, color='#c9520c', lw=2, marker='o', ls='--') # We can see which symbol is which marker from the documentation.
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,x, color='#c9520c', lw=2, marker='o', ls='--', markersize=10)
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,x, color='#c9520c', lw=2, marker='o', ls='--', markersize=10, markerfacecolor='red', markeredgewidth=4, markeredgecolor='orange')
# markerfacecolor specifies the color of the inside of the marker, 
# markeredgecolor specifies the color of the border (frame) of the marker.
```
