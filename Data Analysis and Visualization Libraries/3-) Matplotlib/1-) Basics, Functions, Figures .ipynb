```python
a = np.linspace(0, 10, 11)
# Divide into 11 equal parts starting from 0
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
# The first two values (0, 0) specify the coordinates of the lower left corner of the axes, 
# and the last two values (1, 1) specify its width and height.
# It is specified as a percentage, that is, 1 is the maximum value.
axes.plot(x, y);
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(a, b);
```

```python
fig = plt.figure()

# Large Axes:

axes1 = fig.add_axes([0,0,1,1])
axes1.plot(a,b)

# Small Axes
axes2 = fig.add_axes([0.2, 0.5, 0.2, 0.2])
axes2.plot(x,y)

plt.show() # ";"
```

```python
fig = plt.figure()

# Large Axes:

axes1 = fig.add_axes([0,0,1,1])
axes1.plot(a,b)

axes1.set_xlim(0,8)
axes1.set_ylim(0,8000)
axes1.set_xlabel('A')
axes1.set_ylabel('B')
axes1.set_title('Power of 4')

# Small Axes
axes2 = fig.add_axes([0.2, 0.5, 0.2, 0.2])
axes2.plot(x,y)

axes2.set_xlim(1,2)
axes2.set_ylim(0,5)
axes2.set_xlabel('A')
axes2.set_ylabel('B')
axes2.set_title('Zoomed In')

plt.show() 
```

```python
fig = plt.figure(dpi=200, figsize = (2,2))

axes1 = fig.add_axes([0,0,1,1])
axes1.plot(a,b)

fig.savefig('new_figure.png', bbox_inches = 'tight') # bounding box parameter = 'tight' --> automatically fit x and y axes tick numbers.
```
